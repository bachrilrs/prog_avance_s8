# Spécification des fonctions d’interface
Les opérations autorisées seront accessibles via les seules fonctions suivantes :


**1.   Opérations de création**
*   creer_arbre_vide() : retourne un arbre vide.
*   creer_arbre(r,g,d) : retourne l’arbre constitué de la racine r
et des sous-arbres gauche g et droit d.

**2.   Opérations d’accès**
*   racine(arbre) : retourne la valeur racine de l’arbre passé en
paramètre.
*   gauche(arbre) et droite(arbre) : retournent respectivement
les sous-arbres gauche et droit de l’arbre passé en paramètre.
*   est_vide(arbre) : retourne True si l’arbre passé en paramètre
est vide, False sinon.









```python
def creer_arbre_vide():
    return None

def creer_arbre(r,g,d):
    return r,g,d

def racine(arbre):
    return arbre[0]

def gauche(arbre):
    return arbre[1]

def droite(arbre):
    return arbre[2]

def est_vide(arbre):
    return arbre==None

```

## Parcours en profondeur
Écrire les fonctions récursives **parcours_prefixe**, **parcours_infixe** et **parcours_suffixe** qui acceptent en paramètre un arbre et affichent les valeurs des noeuds de cet arbre selon un parcours en profondeur préfixe, infixe et suffixe.

```python
def parcours_prefixe(tree):
    if est_vide(tree):
        return []
    else:        
        return [racine(tree)] + parcours_prefixe(gauche(tree)) + parcours_prefixe(droite(tree))

print(parcours_prefixe(tree))

def parcours_infixe(tree):
    if est_vide(tree):
        return []
    else:        
        return parcours_infixe(gauche(tree)) + [racine(tree)] + parcours_infixe(droite(tree))

print(parcours_infixe(tree))

def parcours_suffixe(tree):
    if est_vide(tree):
        return []
    else:        
        return parcours_suffixe(gauche(tree))  +  parcours_suffixe(droite(tree)) + [racine(tree)]

print(parcours_suffixe(tree))

```
