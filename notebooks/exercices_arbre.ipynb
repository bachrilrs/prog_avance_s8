# Spécification des fonctions d’interface
Les opérations autorisées seront accessibles via les seules fonctions suivantes :


**1.   Opérations de création**
*   creer_arbre_vide() : retourne un arbre vide.
*   creer_arbre(r,g,d) : retourne l’arbre constitué de la racine r
et des sous-arbres gauche g et droit d.

**2.   Opérations d’accès**
*   racine(arbre) : retourne la valeur racine de l’arbre passé en
paramètre.
*   gauche(arbre) et droite(arbre) : retournent respectivement
les sous-arbres gauche et droit de l’arbre passé en paramètre.
*   est_vide(arbre) : retourne True si l’arbre passé en paramètre
est vide, False sinon.









```python
def creer_arbre_vide():
    return None

def creer_arbre(r,g,d):
    return r,g,d

def racine(arbre):
    return arbre[0]

def gauche(arbre):
    return arbre[1]

def droite(arbre):
    return arbre[2]

def est_vide(arbre):
    return arbre==None
```

## Parcours en profondeur
Écrire les fonctions récursives **parcours_prefixe**, **parcours_infixe** et **parcours_suffixe** qui acceptent en paramètre un arbre et affichent les valeurs des noeuds de cet arbre selon un parcours en profondeur préfixe, infixe et suffixe.

```python
def parcours_prefixe(tree):
    if est_vide(tree):
        return []
    else:        
        return [racine(tree)] + parcours_prefixe(gauche(tree)) + parcours_prefixe(droite(tree))

print(parcours_prefixe(tree))

def parcours_infixe(tree):
    if est_vide(tree):
        return []
    else:        
        return parcours_infixe(gauche(tree)) + [racine(tree)] + parcours_infixe(droite(tree))

print(parcours_infixe(tree))

def parcours_suffixe(tree):
    if est_vide(tree):
        return []
    else:        
        return parcours_suffixe(gauche(tree))  +  parcours_suffixe(droite(tree)) + [racine(tree)]

print(parcours_suffixe(tree))

```

## Parcours en largeur
On souhaite écrire la fonction **parcours_largeur** qui accepte en paramètre un arbre et affiche la valeur des noeuds de cet arbre selon un parcours en largeur.

Contrairement à ce que l’on pourrait croire, cette fonction n’est pas récursive.

L’écriture d’une fonction de parcours en largeur nécessite une structure de données file. Pour l’instant nous utilisons une liste native Python (de type
list) contenant des arbres :
1.   Les arbres sont ajoutés en fin : méthode append(...)
2.   Les arbres à traiter sont retirés en début : méthode .pop(0)

```python
def parcours_largeur(tree):
    bfs = []
    queue = []
    queue.append(tree) # ajoute de l'arbre entier, puis on traite les éléments un par un, on aura les fils de l'arbre dans la queue, et on ajoute les racines à bfs
    while queue: # tant que pas vide
        cur = queue.pop(0) # retire l'arbre 
        if not est_vide(cur): # si pas vide
            queue.append(gauche(cur)) # ajouter la partie gauche pour la traiter plus
            queue.append(droite(cur)) # ajouter la partie droite pour la traiter plus
            bfs.append(racine(cur)) # ajouter la racine
    return bfs

print(parcours_largeur(tree))
print(tree)  
tree[0]
        
```
