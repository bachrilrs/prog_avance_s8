```python
class Abonne:

    def __init__(self):
        pass

    def actualiserEtat(self,capteur):
        """Actualiser l'état du capteur."""
        print(f"Abonne a reçu une notification du capteur {capteur.nom} avec les données : {capteur.get_data()}")
        

class Capteur:
    """Classe capteur."""
    def __init__(self, nom):
        self.nom = nom
        self.liste_diff = []

    def __str__(self):
        return f"{self.nom} avec liste {self.liste_diff}" 
    
    def ajouterAbonne(self,Abonne):
        if Abonne not in self.liste_diff:
            self.liste_diff.append(Abonne)
            print(f"Abonne {self.nom} a été ajouté à la liste de diffusion.")
        else:
            print(f"Abonne {self.nom} est déjà dans la liste de diffusion.")

    def envoyerNotif(self):
        for abonne in self.liste_diff:
            abonne.actualiserEtat(self)
```

```python
class CapteurLuminosite(Capteur):
    """Classe représentant un capteur de luminosité."""
    def __init__(self, nom,intensite):
        super().__init__(nom)
        self.setData(intensite)

    def __str__(self):
        return f"CapteurLuminosite(nom={self.nom}, intensite={self.intensite})"

    def setData(self, intensite : int):
        self.intensite : int = intensite
        self.envoyerNotif()

    def get_data(self):
        return f"Intensité de la luminosité : {self.intensite}"


class CapteurAir(Capteur):
    """Classe représentant un capteur d'air."""

    def __init__(self,nom):
        super().__init__(nom)
        self.temperature = None
        self.humidite = None
        self.co2 = None

    def __str__(self):
        return f"CapteurAire(nom={self.nom}, temperature={self.temperature}, humidite={self.humidite}, co2={self.co2})"
    
    def setData(self, temperature : float, humidite : float , co2 : float):
        self.temperature= temperature
        self.humidite = humidite
        self.co2 = co2
        super().envoyerNotif()
    
    def get_data(self):
        return f"Degré Celsius : {self.temperature}, Humidité : {self.humidite}, Taux de co2 : {self.co2}"
    

```

```python
class StationMeteo(Abonne):
    """Station météo"""

    def __init__(self,nom):
        super().__init__()
        self.nom = nom

    def __str__(self):
        return f"Station météo {self.nom}"
    
    def actualiserEtat(self,capteur):
        print(f"Station a recu mise à jour de {capteur.nom}.")

        if hasattr(capteur,"intensite"):
            print(f"Intensité du capteur : {capteur.intensite}")
    
```

```python
class VoletRoulant(Abonne):
    """Classe représentant un volet roulant."""
    def __init__(self,open : bool):
        self.opened = open

    def __str__(self):
        return f"VoletRoulant(opened={self.opened})"
    
    def actualiserEtat(self,capteur):

        if isinstance(capteur,CapteurLuminosite):
            if capteur.intensite > 100 and self.opened:
                self.opened = False
                print(f"Volets ouvert : {self.opened}")

            if capteur.intensite < 20 and not self.opened:
                self.opened = True
                print(f"Volets ouvert : {self.opened}")
        else:
            print(f"Ce n'est pas un capteur de luminosité")
        
```
