```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("../data/Athletes_summer_games.csv")

df.head()
```

```python
df.dropna(inplace=True)
print(df.columns)
df.Age.astype(int).mean()

try:
    df.drop('Unnamed: 0',axis='columns',inplace=True)
except KeyError:
    print('Colonne pas présente.')

df.groupby('Team').describe()
```

```python
df_sport_team = df.groupby(['Team','Sport'])['Age'].mean() # series one dimension
df_sport_team = df.groupby(['Team','Sport'])[['Age']].mean() # df 
df_sport_team['x'] = df.groupby(['Team','Sport'])['Age'].mean() - 2000 # create a column

# When you run a groupby(['Team', 'Sport']), Pandas takes those two columns and moves them into the Index (the row labels). 
# They are no longer considered "Data Columns."
# So shape (...,1)

arbres = df.loc[df.Team.isin(['Algeria','Morocco']) ]

arbres
```
