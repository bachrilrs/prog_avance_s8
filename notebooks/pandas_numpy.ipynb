```python
# print(df.groupby('dep_code'))


# for key, item in grouped_df:
#     print(grouped_df.get_group(key), "\n\n")


cols = set(df.columns)

needed_cols = ['code_insee'
,'nom_sans_accent'
,'reg_nom'
,'dep_code'
,'academie_nom'
,'type_commune_unite_urbaine'
,'population'
,'superficie_hectare'
,'altitude_moyenne'
,'latitude_centre'
,'longitude_centre'
,'niveau_equipements_services']


df_2 = df[['code_insee','nom_sans_accent','reg_nom','dep_code','academie_nom','type_commune_unite_urbaine','population','superficie_hectare','altitude_moyenne','latitude_centre','longitude_centre','niveau_equipements_services']]

df_2 = df.loc[:,['code_insee','nom_sans_accent','reg_nom','dep_code','academie_nom','type_commune_unite_urbaine','population','superficie_hectare','altitude_moyenne','latitude_centre','longitude_centre','niveau_equipements_services']]


df_2.head()


print(df.population.astype(int).sum())


# print(df[df.loc[:,'nom_sans_accent']=='moissac'])


# print(df.loc[(df.population.astype(int)>100000)])

df_100k = df[df.population.astype(int)>100000]


df[df.population.astype(int)>10000].groupby(['reg_nom']).size().sort_values(ascending=False).head(3)

# Quelle est la ville la plus proche du centre de la France Métropolitaine ?


df['latitude_centre'] = df.latitude_centre.astype(float)
df['longitude_centre'] = df.longitude_centre.astype(float)


df['distance_centre'] = np.sqrt((df.latitude_centre-46.603354)**2 + (df.longitude_centre-1.888334)**2)
# explication : distance entre 2 points dans un plan : sqrt((x2-x1)^2 + (y2-y1)^2) 
# df.latitude_centre-46.603354 : distance en latitude entre la ville et le centre de la France
# df.longitude_centre-1.888334 : distance en longitude entre la ville et le centre de la France
# on calcule la distance euclidienne entre la ville et le centre de la France
# on ajoute une nouvelle colonne 'distance_centre' au dataframe df qui contient la distance entre chaque ville et le centre de la France
# on utilise la fonction np.sqrt() pour calculer la racine carrée de la somme des carrés des distances en latitude et en longitude
# on utilise la fonction sort_values() pour trier le dataframe df par la colonne 'distance_centre' et on affiche la première ligne du dataframe trié qui correspond à la ville la plus proche du centre de la France
# comment t'as obtenu les coordonnées du centre de la France ? : j'ai utilisé les coordonnées géographiques du centre de la France Métropolitaine qui sont latitude 46.603354 et longitude 1.888334
df.sort_values(by='distance_centre').head(1)
```
