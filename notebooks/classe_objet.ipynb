```python
class Vecteur:
    "Classe vecteur"

    def __init__(self, x,y,z):

        self.__x = x
        self.__y = y
        self.__z = z
        self.calculer_norme()


    def __str__(self):
        return f'Vecteur avec direction {(self.__x,self.__y,self.__z)} avec norme {self.__norme}'
    

    @classmethod
    def creer_vecteur(cls, p1: Point, p2: Point):
        """Creer vecteur avec 2 points"""
        return cls( p1.x-p2.x, p1.y-p2.y, p1.z-p2.z) # va créer un vecteur qui va de p2 vers p1
    
    def calculer_norme(self):
        """Calcule norme"""
        self.__norme = np.sqrt((self.__x**2) + (self.__y**2) + (self.__z**2))
    
    def __add__(self,other):
        return Vecteur(self.__x + other.__x , self.__y + other.__y, self.__z + other.__z)

    def __sub__(self, other):
        return Vecteur(self.__x - other.__x , self.__y - other.__y, self.__z - other.__z)

    def __mul__(self,other):
        return Vecteur(self.__x * other.__x , self.__y * other.__y, self.__z * other.__z)
    
    def __eq__(self,other):
        return self.calculer_norme() == other.norme()
    
    def __ne__(self,other):
        return True if not self.__eq__(other) else False

    def produit_vectoriel(self,  vect : 'Vecteur') -> 'Vecteur':
        """Retourne le produit vectoriel"""
        dx = self.__y*vect.__z - self.__z*vect.__y
        dy = self.__z*vect.__x - self.__x*vect.__z
        dz = self.__x*vect.__y - self.__y*vect.__x
        return Vecteur( dx,dy,dz)
    
    @property
    def x(self):
        return self.__x
    
    @property
    def y(self):
        return self.__y
    
    @property
    def z(self):
        return self.__z
    
    @property
    def norme(self):
        return self.__norme

    @x.setter
    def x(self,x):
        """assurer que chaque fois quon modifie le x,y,z on calcule la norme"""
        self.__x = x
        self.calculer_norme()
    
    @y.setter
    def y(self,y):
        self.__y = y
        self.calculer_norme()
    
    @z.setter
    def z(self,z):
        self.__z = z
        self.calculer_norme()
    

    
p1 = Point('p1',1,2,3)
p2 = Point('p2',4,6,7)
p3 = Point('p3',4,9,7)
p4 = Point('p4',2,6,5)

v1_2 = Vecteur.creer_vecteur(p1,p2)
v3_4 = Vecteur.creer_vecteur(p3,p4)

v2 = Vecteur(3, 8, 9)

produit_vect = v1_2.produit_vectoriel(v3_4)

print(v1_2)
print(v3_4)

print(v1_2 + v3_4)

```

```python
class Triangle:
    "Classe triangle"

    def __init__(self, a: Point, b: Point, c: Point):
        self.a = a
        self.b = b
        self.c = c

    def __str__(self):
        return f'Points {self.a}, {self.b}, {self.c}'
    
    def surface(self)->float:
        vectAB = Vecteur.creer_vecteur(self.a , self.b)
        vectAC = Vecteur.creer_vecteur(self.a , self.c)

        prod = vectAB * (vectAC)
        print(type(prod))
        print(prod)
        s = 1/2 * prod.norme

        return s
    
    def est_rectangle(self) -> bool:
        """Retourne True si le triangle est rectangle"""
        vectAB = Vecteur.creer_vecteur(self.a , self.b)
        vectAC = Vecteur.creer_vecteur(self.a , self.c)
        vectBC = Vecteur.creer_vecteur(self.b , self.c)

        AB2 = vectAB.norme**2
        AC2 = vectAC.norme**2
        BC2 = vectBC.norme**2

        return (AB2 + AC2 == BC2) or (AB2 + BC2 == AC2) or (AC2 + BC2 == AB2)
    

t1 = Triangle(p1,p2,p3)
print(t1)

print(t1.surface())
print(t1.est_rectangle())
```

```python
# Exo 1
from enum import Enum, auto

class NomMatiere(Enum):
    """Matière"""
    MATHS = auto()
    FRANCAIS = auto()
    ANGLAIS = auto()

    def __str__(self):
        return super().__str__()
    
    
class Controle:
    """Controle avec notes et coefficient"""

    def __init__(self, nom:str, note= np.nan, coefficient=1):
        self.nom = nom
        self.note = note
        self.coefficient = coefficient

    def __str__(self):
        return f'Note {self.note} avec Coefficient {self.coefficient}'
 

class Matiere:
    """Notes"""

    def __init__(self,NomMatiere):
        self.notes : list= []
        self.NomMatiere = NomMatiere
        self.moyenne = self.calculerMoyenne() # calcule automatiquement la moyenne à chaque ajout de note

    def __str__(self):
        return f'{self.moyenne}'

    def ajouterNote(self, Controle):
        """Ajouter Controle dans la liste de Controle"""
        self.notes.append(Controle)
        self.calculerMoyenne() # recalcule matiere
        return f'{Controle} a ete ajouté aux Notes'
    
    def calculerMoyenne(self):
        """Calculer moyenne prenant en compte le coeff"""
        moyenne = 0
        somme_c= 0
        if len(self.notes):
            for n in self.notes:
                somme_c += n.coefficient
                moyenne += n.note*n.coefficient
            self.moyenne = moyenne/somme_c
        else:
            self.moyenne = 0
        return self.moyenne
```

```python
class Eleve:
    """Eleve"""

    def __init__(self, nom, prenom):
        self.nom = nom
        self.prenom = prenom
        self.notes_eleve= {}

        for mat in NomMatiere:
            self.notes_eleve[mat]= Matiere(mat)

    def __str__(self):
        return f'{self.nom} {self.prenom}'

    def afficher_notes(self):
        for v in self.notes_eleve.values():
            print(f'{v.NomMatiere} : Moyenne : {v.moyenne}')
        return 

    def ajouterNote(self, Controle:Controle, nmat:Matiere):
        self.notes_eleve[nmat].ajouterNote(Controle)

    def notemoyMatiere(self, Matiere):
        return Matiere.moyenne
    
    def moyenne_generale(self):
        n = len(self.notes_eleve)
        somme = 0
        for v in self.notes_eleve.values():
            somme += int(v.moyenne)

        return somme / n 
    
    def afficherNoteEleve_Matiere(self,Eleve, NomMatiere):
        return f"{Eleve} a {Eleve.notes_eleve[NomMatiere].moyenne:0.2f} de moyenne en {NomMatiere.name.lower()}"

        
```

```python
class Classe:
    """Classe"""

    def __init__(self,nom: str):
        self.nom = nom
        self.eleves = {}

    def __str__(self):
        return f'La classe {self.nom} est composé de {len(self.eleves.keys())}'
    
    def ajouterEleve(self, Eleve: Eleve):
        if Eleve not in self.eleves:
            self.eleves[Eleve.__str__()] = Eleve

    def ajouterNoteEleveMatiere(self, Eleve, Controle, NomMatiere):
        Eleve.ajouterNote(Controle, NomMatiere)

    def bulletinEleve(self, Eleve):
        print(f"\n\nBulletin de {Eleve.nom} {Eleve.prenom} :")
        for matiere, note in Eleve.notes_eleve.items():
            print(f"{matiere.name} : {note.moyenne}")
                
    def notesMatiere(self, Eleve, nom_matiere : NomMatiere):
        """Retourne la note d'un eleve pour une matiere donnée"""
        print(f'Notes des {Eleve.__str__()} en {nom_matiere}')
        return Eleve.notes_eleve[nom_matiere]
    
    def noteMatiere_Classe(self, NomMatiere):
        '''Retourne la note de tous les eleves dans une matiere'''
        print(f'\n\n\nNotes de la classe en {NomMatiere.name.lower()} :')
        for v in self.eleves.values():
            print(f'{v} : {v.notes_eleve[NomMatiere].moyenne}')
    
```

```python
class Node:
    """Node pour arbre binaire"""

    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

    def __str__(self):
        return f'{self.left if self.left else ""} {self.value} {self.right if self.right else ""}'

    def add_value(self,value):

        if value>self.value:
            if self.right:
                self.right.add_value(value)
            else:
                self.right = Node(value)

                # plus grand a droite
        
        else:
            if self.left:
                self.left.add_value(value)
            else:
                self.left = Node(value)

                # plus petit a gauche

    def rechercher_valeur(self, valeur):
        if self.value == valeur:
            return True
        
        if valeur>self.value:
            if self.left is not None:
                return self.left.rechercher_valeur(valeur)
            else:
                return False
        else:
            if self.right is not None:
                return self.right.rechercher_valeur(valeur)
            else:
                return False

    def count_nodes(self):

        total = 0

        if self.left is not None:
            total += self.left.count_nodes()
        
        if self.right is not None:
            total += self.right.count_nodes()

        return 1 + total
    
    def hauteur(self):
        if self.left is None and self.right is None:
            return 1
        elif self.left is None:
            return 1 + self.right.hauteur()
        elif self.right is None:
            return 1 + self.left.hauteur()
        else:
            return 1 + max(self.left.hauteur(), self.right.hauteur())


n1 = Node(54)
n1.add_value(4)
n1.add_value(60)
n1.add_value(78)
n1.add_value(4)
n1.add_value(4)



print(n1.rechercher_valeur(78))

n1.count_nodes()
n1.hauteur()
n1.__str__()

```

```python
class Personne:
    """personne"""

    def __init__(self,nom,prenom,age):
        self.__nom = nom
        self.__prenom = prenom
        self.__age = age

    def __str__(self):
        return f"{self.__nom} {self.__prenom} is {self.__age} years old."

    def __eq__(self, other):
        if self.nom == other.nom:
            return True
        return False
    
    def __gt__(self, other):

        if self.nom > other.nom:
            return True
        else:
            return False
        
    def __lt__(self, other):

        if self.nom < other.nom:
            return True
        else:
            return False
    

    @property
    def nom(self):
        return self.__nom
    
    @property
    def prenom(self):   
        return self.__prenom
    
    @property
    def age(self):
        return self.__age

    @nom.setter
    def nom(self,nom):
        self.__nom = nom

    @prenom.setter
    def prenom(self,prenom):
        self.__prenom = prenom

    @age.setter
    def age(self,age):
        if age < 0:
            raise ValueError("Age can't be negative")
        self.__age = age

milo = Personne("georges","milo",52)
loic = Personne("georges","loic",59)
gwen = Personne("georges","gwen",26)
stephane = Personne("georges","stephane",59)
thalie = Personne("georges","thalie",26)
samina = Personne("georges","samina",52)




milo.__eq__(loic)


bbs = Node(loic)



for p in [milo, gwen, stephane, thalie, samina]:
    bbs.add_value(p)

bbs.__str__()
```

```python
class ArbreLexical:
    """arbre lexical"""

    def __init__(self, poids):
        self.poids = poids
        self.enfants = {}


    def ajouter_mot(self, mot):
        current_node = self
        for lettre in mot:
            if lettre not in current_node.enfants:
                current_node.enfants[lettre] = ArbreLexical(0)
            current_node = current_node.enfants[lettre]
        current_node.poids += 1
      

    def __str__(self):
        result = f'Poids: {self.poids}\n'
        for lettre, enfant in self.enfants.items():
            result += f'Lettre: {lettre} : {enfant.__str__()}'
        return result

    def count_letter(self, mot):
        self.count = Counter(mot)


mot = "georges"
arbre = ArbreLexical(0)

arbre.ajouter_mot("milo")
arbre.ajouter_mot("mila")
arbre.ajouter_mot("mile")
arbre.ajouter_mot("mili")
arbre.ajouter_mot("milu")
arbre.ajouter_mot("milua")
arbre.ajouter_mot("miluo")

print(arbre)
```
