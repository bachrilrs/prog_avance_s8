```python
def insert_sort(liste):
    n = len(liste)

    end = n 
    start = 0
    for i in range(start + 1, end):
        k = liste[i]
        j = i 
        while j > start and liste[j-1] > k:
            liste[j] = liste[j-1]
            j -= 1
        liste[j] = k
    return liste
```

```python
def quicksort(liste):
    if len(liste) <= 1:
        return liste
    n = len(liste)
    centre = n//2

    pivot = liste[centre]

    gauche = [e for e in liste if e<pivot]
    middle = [e for e in liste if e == pivot]
    droite = [e for e in liste if e>pivot]

    return quicksort(gauche) + middle + quicksort(droite)
```

*   un entier $ancienne\_valeur$,
*   un entier $nouvelle\_valeur$.

La fonction retourne une nouvelle liste dans laquelle la première occurrence de
$ancienne\_valeur$ est remplacée par $nouvelle\_valeur$.


```python
def substituer_premiere_occ(liste, old, new):

    if est_vide(liste):
        return creer_liste_vide()
    if tete(liste) == old:
        return creer_liste(new, queue(liste))
    else:
        return creer_liste(tete(liste), substituer_premiere_occ(queue(liste), old, new))
    

print(substituer_premiere_occ(liste, 59, 42))
print(longueur(liste))
```

```python
def creer_arbre_vide():
    return None

def creer_arbre(r,g,d):
    return r,g,d

def racine(arbre):
    return arbre[0]

def gauche(arbre):
    return arbre[1]

def droite(arbre):
    return arbre[2]

def est_vide(arbre):
    return arbre==None
```

```python
tree = (8, (21, (34, None, None), (18, (16, None, None), (19,None, None))), (6, (13, None, None), None)) 
```

```python
def calculer_taille(tree):
    if est_vide(tree):
        return 0 
    else:
        return 1 + calculer_taille(gauche(tree)) + calculer_taille(droite(tree))
    

def hauteur_arbre(tree):
    if est_vide(tree):
        return 0
    else:
        return 1 + max(hauteur_arbre(gauche(tree)), hauteur_arbre(droite(tree)))


print(tree)
print(calculer_taille(tree))
print(hauteur_arbre(tree))

```

## Parcours en profondeur
Écrire les fonctions récursives **parcours_prefixe**, **parcours_infixe** et **parcours_suffixe** qui acceptent en paramètre un arbre et affichent les valeurs des noeuds de cet arbre selon un parcours en profondeur préfixe, infixe et suffixe.

```python
def parcours_prefixe(tree):
    if est_vide(tree):
        return []
    else:
        return [racine(tree)] + parcours_prefixe(gauche(tree)) + parcours_prefixe(droite(tree))
    
print(parcours_prefixe(tree))

def parcours_infixe(tree):
    if est_vide(tree):
        return []
    else:
        return parcours_infixe(gauche(tree)) + [racine(tree)] +  parcours_infixe(droite(tree))

print(parcours_infixe(tree))


def parcours_suffixe(tree):
    if est_vide(tree):
        return []
    else:
        return parcours_suffixe(gauche(tree)) +  parcours_suffixe(droite(tree)) + [racine(tree)] 

print(parcours_suffixe(tree))
```

## Parcours en largeur
On souhaite écrire la fonction **parcours_largeur** qui accepte en paramètre un arbre et affiche la valeur des noeuds de cet arbre selon un parcours en largeur.

Contrairement à ce que l’on pourrait croire, cette fonction n’est pas récursive.

L’écriture d’une fonction de parcours en largeur nécessite une structure de données file. Pour l’instant nous utilisons une liste native Python (de type
list) contenant des arbres :
1.   Les arbres sont ajoutés en fin : méthode append(...)
2.   Les arbres à traiter sont retirés en début : méthode .pop(0)

```python
def bfs(tree):

    bfs = []
    queue = []

    queue.append(tree)

    while queue:
        cur = queue.pop(0)
        if not est_vide(cur):
            queue.append(gauche(cur))
            queue.append(droite(cur))
            bfs.append(racine(cur))
    return bfs

print(bfs(tree))
```

La construction de l’arbre s’effectue de la manière suivante :
Le premier élément de cet ensemble (50) servira de racine à notre arbre,
- Les éléments suivants sont ensuite ajoutés par ordre d’apparition dans l’ensemble.
Pour ajouter une nouvelle valeur, on part de la racine de l’arbre, en considérant la racine comme
le nœud courant. On peut ajouter un élément au nœud courant, si


l’élément à ajouter est strictement inférieur à l’élément du nœud courant
- si ce dernier n’a pas de fils gauche, on crée un nouveau nœud avec cette valeur
et on l’associe au fils gauche ;

- si le nœud courant a déjà un fils gauche, on essaye alors d’insérer cette nouvelle
valeur à partir du fils gauche ;


 l’élément à ajouter est supérieur à l’élément du nœud courant
- si ce dernier n’a pas de fils droit, on crée un nouveau nœud avec cette valeur
et on l’associe au fils droit ;
- si le nœud courant a déjà un fils droit, on essaye alors d’insérer cette nouvelle
valeur à partir du fils droit ;

```python


def add_value(tree, v):

    if est_vide(tree):
        return creer_arbre(v, creer_arbre_vide(), creer_arbre_vide())

    r = racine(tree)
    g = gauche(tree)
    d = droite(tree)

    if v < r:
        # On reconstruit l'arbre avec le nouveau sous-arbre gauche
        return creer_arbre(r, add_value(g, v), d)
    else:
        # On reconstruit l'arbre avec le nouveau sous-arbre droit
        return creer_arbre(r, g, add_value(d, v))
    

tree = creer_arbre_vide()

tree = add_value(tree,5)
tree = add_value(tree,64)
tree = add_value(tree,2)

tree = add_value(tree,9)
tree = add_value(tree,99)
tree = add_value(tree,45)


print(tree)
parcours_infixe(tree)
```
