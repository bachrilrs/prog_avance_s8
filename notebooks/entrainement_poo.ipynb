```python
class Dispositif:
    nb_dispositifs = 0
    def __init__(self, id,nom):
        self.__id = id
        self.__nom = nom
        Dispositif.nb_dispositifs += 1

    def __str__(self):
        return f"Dispositif {self.__id} : Nom : {self.__nom} "
    
    def traiter_donnes(self):
        return f"Traitement générique du dispositif {self.__id}"

    @property
    def id(self):
        return self.__id
    
    @id.setter
    def id(self,id):
        self.__id = id

    @property
    def nom(self):
        return self.__nom
    
    @nom.setter
    def nom(self,nom):
        self.__nom = nom

class Capteur(Dispositif):

    def __init__(self,id,nom,v):
        super().__init__(id,nom)
        self.__v = v
    
    @property
    def v(self):
        return self.__v
    
    @v.setter
    def v(self,v):
        self.__v = v

    
    def __str__(self):
        return super().__str__() + f"Valeur : {self.__v}"
    
    def traiter_donnes(self,seuil=50):
        if self.__v > seuil:
            return f"ALERTE : Valeur critique sur [{self.__nom} !"

class Actionneur(Dispositif):

    def __init__(self, id, nom, etat=1):
        super().__init__(id, nom)
        self.__etat = etat
    
    @property
    def etat(self):
        return self.__etat
    
    @etat.setter
    def etat(self,etat):
        self.__etat = etat

    def __str__(self):
        return super().__str__() + f"Etat : {self.__etat}"
    
    def traiter_donnes(self):
        if self.__etat:
            self.etat = 0
        else:
            self.etat = 1

 
a = Actionneur(3, 'fdf' , etat=1)
c = Capteur(1, 'capteur1', v=60)  
f = Dispositif(2, 'dispositif1')

print(Dispositif.nb_dispositifs)
```

```python
# Exo 2

class Document:

    def __init__(self,title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f"Document {self.title} written by {self.author}"
    
class Livre(Document):

    def __init__(self, title : str, author : str, nb_pages: int):
        super().__init__(title, author)
        self.nb_pages = nb_pages

    def __str__(self):
        return super().__str__() + f" {self.nb_pages}"
    

class Ebook(Livre):

    def __init__(self, title : str, author : str, nb_pages: int , format : str):
        super().__init__(title, author, nb_pages)
        self.format = format

    def __str__(self):
        return super().__str__() + f" Format : {self.format}"
    

d = Document('ttest ', "don B")
print(d)
l = Livre('ttest ', "don B", 56)
print(l)

e = Ebook('ttest ', "don B", 56, 'pdf')
print(e)
```

```python
# exo 1
class Coffrefort:

    def __init__(self, code : int, proprietaire : str ):
        self.__code = code
        self._proprietaire = proprietaire 

    @property
    def proprietaire(self):
        return self._proprietaire
    
    @property
    def code(self):
        return self.__code
    
    def check_code(self,try_code):
        return True if self.__code == try_code else False
    
    

class CoffreBanque(Coffrefort):

    def __init__(self, code, proprietaire, frais_ouverture):
        super().__init__(code, proprietaire)
        self.__frais_ouverture = frais_ouverture

    def __str__(self):
        return f"Proprietaire : {self._proprietaire} Frais ouverture : {self.__frais_ouverture}"
    
    @property
    def frais_ouverture(self):
        return self.__frais_ouverture
    
    
    

c = Coffrefort(123,'Don b')
c.__str__()

cb = CoffreBanque(123,'Don b',4)

cb.check_code(133)
```

```python
def creer_arbre_vide(): 
    return None

def est_vide(a): 
    return a is None

def racine(a): 
    return a[0]

def gauche(a): 
    return a[1]

def droite(a): 
    return a[2]

def creer_arbre(r, g, d): 
    return (r, g, d)
```

```python
# exo 2
class Employee:

    def __init__(self, name : str , salary: float):
        self.name = name
        if salary < 0:
            raise ValueError("Salary cannot be negative")
        if type(salary) not in [int, float]:
            raise TypeError("Salary must be a number")
        self.salary = salary

    def compute_salary(self):
        return self.salary
    
    

class Commercial(Employee):

    def __init__(self, name, salary, commission):
        super().__init__(name, salary)
        self.commission = commission

    def compute_salary(self):
        return super().compute_salary() + self.commission

class Director(Employee):

    def __init__(self, name, salary, bonus_fixe):
        super().__init__(name, salary)
        self.bonus_fixe = bonus_fixe
    
    def compute_salary(self):
        return super().compute_salary() + self.bonus_fixe



def add_employee(tree, emp):

    if est_vide(tree):
        return creer_arbre(emp, creer_arbre_vide(), creer_arbre_vide())

    if emp.salary < racine(tree).salary:
        return creer_arbre(racine(tree), add_employee(gauche(tree),emp), droite(tree))
    else:
        return creer_arbre(racine(tree), gauche(tree), add_employee(droite(tree),emp))

def parcours_infixe_salaries(tree):
    if est_vide(tree):
        return []
    return parcours_infixe_salaries(gauche(tree)) + [racine(tree).compute_salary()] + parcours_infixe_salaries(droite(tree))

def all_salaries(tree):
    return sum(parcours_infixe_salaries(tree))

def all_salaries_2(tree):
    if est_vide(tree):
        return 0
    return all_salaries_2(gauche(tree)) + racine(tree).compute_salary() + all_salaries_2(droite(tree))


# generate  employees

e1 = Employee("Alice", 50000)
e2 = Commercial("Bob", 40000, 5000)
e3 = Director("Charlie", 60000, 10000)
tree = creer_arbre_vide()
tree = add_employee(tree, e1)
tree = add_employee(tree, e2)
tree = add_employee(tree, e3)

print(parcours_infixe_salaries(tree))

print(all_salaries(tree))
print(all_salaries_2(tree))


```
